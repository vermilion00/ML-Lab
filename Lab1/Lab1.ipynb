```python
import numpy as np
import matplotlib.pyplot as plt

#Seed the rng function
rng = np.random.default_rng(1234567)

#standard floats
standard = rng.random(size=500) * 7

#normal floats
normal = rng.standard_normal(500) + 3.5

#Show first histogram
plt.title('Standard float distribution, 0-7')
plt.hist(standard, bins=500, alpha=1)
plt.show()

#Show second histogram
plt.title('Normal float distribution, 0-7')
plt.hist(normal, bins=500, alpha=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

#Read csv file, label columns correctly
df = pd.read_csv('regression.csv', header=0, names=['Years of Experience', 'Salary'])

#Split dataset into x and y sets by column label
x = df['Years of Experience']
y = df['Salary']

#Create scatter Plot
#plt.scatter(x, y, label='Original Data')
plt.plot(x, y, 'o', label='Scatter plot')


#Linear regression of x and y sets
res = stats.linregress(x, y)

#Create linear regression on top of scatter plot
plt.plot(x, res.intercept + res.slope*x, 'r', label='Fitted line')
#Draw graph
#Label graph
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()



```
